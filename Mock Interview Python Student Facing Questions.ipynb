<b> Put your code below </b>

```python
# To make a bar graph of the count of 'unique' job roles, first by subsetting for the 'Job-Role' column in the filtered dataframe, then
# using the .value_counts() method to count the values/variables in this column. I then followed it with the sort_values() method to make sure 
# that its in ascending order, which is the default of this method. This is set to a new name. Next, to plot it, I added the .plot()
# method after the updated dataframe name, also adding kind='bar' to specify that it should be a barplot.

#df_filtered['Job-Role'].sort_values().plot(kind='bar')
dff_count= df_filtered['Job-Role'].value_counts().sort_values()
dff_count.plot(kind= 'bar')
```

```python
#To plot for these 2 requirements, first I used boolean logic to make an updated dataframe that's specific to only the 2 criteria listed for 
# one graph, with '&' placed between them to specify that both criteria are required, making sure to use parenthesis() to group the 2 
# conditions and brackets[] placed around the column names as well as the Boolean mask. Then to count these job roles, I subset for that 
# Job-Role column in this further filtered dataframe and add the .value_counts() method. Next, I can plot the graph by using the 
# .plot( ) method, adding the kind needed, bar, and the title also, inside these parenthesis. Next I can make it visible by using
# the plt.show() method. This is done twice, so for each graph type.

dff_count_HS_50= df_filtered[(df_filtered['Degree Status']== ' HS-grad')& (df_filtered['Earnings']== ' <=50K') & (df_filtered['Country']== ' United-States')]
#print(df_filtered['Degree Status'].unique())
#print(df_filtered['Earnings'].unique())
#print(count1)
#count1['Job-Role'].plot(kind= 'bar')<-- this is just a number not the rows,so can't plot it!

# Counting job roles (value counts) in that filtered DataFrame
job_counts1 = dff_count_HS_50['Job-Role'].value_counts()
# Plotting the result
job_counts1.plot(kind='bar', title='Individuals who earn <=50K, have a High School Graduate Diploma and are in the United States')
plt.show()

#seem to need plt.show present for each to separate code into 2 graphs..
dff_count_HS__50= df_filtered[(df_filtered['Degree Status']== ' HS-grad') & (df_filtered['Earnings']==' >50K') & (df_filtered['Country']== ' United-States')]
#print(count2)

job_counts2 = dff_count_HS__50['Job-Role'].value_counts()
job_counts2.plot(kind='bar',title= 'Individuals who earn >50K, have a High School Graduate Diploma and are in the United States')
plt.show()
```
